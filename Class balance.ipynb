```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets (replace with actual file paths)
train_df = pd.read_csv('train_data(new2).csv')
validate_df = pd.read_csv('val_data(new2).csv')
test_df = pd.read_csv('test_data(new2).csv')

# List of datasets to process
datasets = {
    "Train Dataset": train_df,
    "Validate Dataset": validate_df,
    "Test Dataset": test_df
}

# Function to analyze each dataset
def analyze_dataset(name, df):
    print(f"\nAnalysis for {name}:\n" + "-"*50)
    
    # Convert date column to datetime
    df['Entry_Date'] = pd.to_datetime(df['Entry_Date'])

    # 1. Count of Target Variables
    target_counts = df['Target'].value_counts()
    print("Count of Target Variables:")
    print(target_counts)
    
    # 2. Timeframe of the dataset
    start_date = df['Entry_Date'].min()
    end_date = df['Entry_Date'].max()
    timeframe = f"{start_date} to {end_date}"
    print("\nTimeframe of the Dataset:")
    print(timeframe)
    
    # 3. Plotting Profit/Loss over time
    plt.figure(figsize=(10, 6))
    plt.plot(df['Entry_Date'], df['Last'], label='Closing Price')
    plt.title(f'Profit/Loss Over Time ({name})')
    plt.xlabel('Date')
    plt.ylabel('Closing price')
    plt.grid(True)
    plt.show()

    # 4. Average Profit/Loss
    avg_profit_loss = df['Profit_Loss'].mean()
    print(f"\nAverage Profit/Loss: {avg_profit_loss:.4f}")

# Analyze each dataset
for name, df in datasets.items():
    analyze_dataset(name, df)
```
